```python
df = df.fillna(df['Arrival Delay in Minutes'].mean)
df.isnull().sum()
```

```python
df.describe(include='all')
```

```python
#Splitting the entire data into train and test data....

from sklearn.model_selection import train_test_split

df['satisfaction'].value_counts()
y=df.replace({'satisfaction':{'neutral or dissatisfied':0,'satisfied':1}})
y=y['satisfaction']
y

```

```python
print(df['Customer Type'].value_counts())
print(df['Type of Travel'].value_counts())
df = df.replace({'Customer Type':{'Loyal Customer':1,'disloyal Customer':0}})
df = df.replace({'Type of Travel':{'Business travel':1,'Personal Travel':0}})
df.head()
```

```python
columns = ['Customer Type', 'Age', 'Type of Travel', 'Flight Distance', 'Inflight wifi service',
       'Departure/Arrival time convenient', 'Ease of Online booking',
       'Gate location', 'Food and drink', 'Online boarding', 'Seat comfort',
       'Inflight entertainment', 'On-board service', 'Leg room service',
       'Baggage handling', 'Checkin service', 'Inflight service',
       'Cleanliness', 'Departure Delay in Minutes']
X = df[columns]
X 
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=23)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

model = DecisionTreeClassifier()
model.fit(X_train,y_train)
prediction = model.predict(X_test)
print(prediction)
accuracy = accuracy_score(y_test,prediction)
accuracy
```

```python
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt

plt.figure(figsize=(20,10))
plot_tree(model, filled=True, feature_names=columns, class_names=['Neutral or Dissatisfied', 'Satisfied'])
plt.show()
```
